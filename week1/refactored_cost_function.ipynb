```python
import numpy as np
import sklearn as sk
import sklearn.datasets
import matplotlib.pyplot as plt
import pandas as pd
```

```python
x, y = sk.datasets.make_regression(n_features=1, noise=10)

plt.plot(x.ravel(), y, 'o')
plt.show()
```

```python
'''
np.sum((theta0 + theta1 * x - y) ** 2) / (2.0 * len(x))
'''
```

```python
t0, t1 = np.meshgrid(np.linspace(-100, 100, 1000), np.linspace(-100, 100, 1000))
t = np.asmatrix(np.row_stack((t0.ravel(), t1.ravel())))

t
```

```python
xm = np.asmatrix(np.column_stack((np.repeat(1, x.shape[0]), x)))

xm
```

```python
ym = np.repeat(y.reshape(y.size, 1), t.shape[1], axis=1)

ym
```

```python
j = np.sum(np.asarray(xm * t - ym) ** 2, axis=0) / (2.0 * x.size)

j
```

```python
df = pd.DataFrame(list(zip(t0.ravel(), t1.ravel(), j)), columns=['theta0', 'theta1', 'cf'])

df.sort_values(['cf'], ascending=True)
```
