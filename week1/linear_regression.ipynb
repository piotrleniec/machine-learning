```python
import matplotlib.pyplot as plt
import sklearn as sk
import sklearn.datasets
from sklearn.linear_model import SGDRegressor
```

```python
x, y = sk.datasets.make_regression(n_features=1, noise=10)

sgdr = SGDRegressor(n_iter=100)
sgdr.fit(x, y)

plt.plot(x.ravel(), y, 'o')
plt.plot(x, sgdr.predict(x.reshape(100, 1)).ravel())
plt.show()
```
