```python
import numpy as np
import sklearn as sk
import sklearn.linear_model
import pandas as pd
import matplotlib.pyplot as plt
```

```python
theta0, theta1 = np.random.uniform(-10, 10, 2)
x = np.random.uniform(1000, 2000, 50)
y = theta0 + theta1 * x + 500 * np.random.randn(50)

df = pd.DataFrame(np.column_stack((x, y)), columns=['x', 'y'])

plt.plot(x, y, 'o')
plt.show()

df.x.describe()
```

```python
sgdr = sk.linear_model.SGDRegressor(n_iter=100)
sgdr.fit(x.reshape(x.size, 1), y)

plt.plot(x, y, 'o')
plt.plot(x.ravel(), sgdr.predict(x.reshape(50, 1)))
plt.show()
```

```python
sx = sk.preprocessing.scale(x)

sgdr = sk.linear_model.SGDRegressor(n_iter=100)
sgdr.fit(sx.reshape(sx.size, 1), y)

plt.plot(sx, y, 'o')
plt.plot(sx.ravel(), sgdr.predict(sx.reshape(50, 1)))
plt.show()
```
