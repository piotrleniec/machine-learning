```python
import numpy as np
import sklearn as sk
import sklearn.datasets
import sklearn.linear_model
import sklearn.metrics
import matplotlib.pyplot as plt
```

```python
x, y = sk.datasets.make_regression(n_features=1, noise=7.5)

plt.plot(x, y, 'o')
plt.show()
```

```python
def regression_error(n_iter, alpha=0.0001):
    sgdr = sk.linear_model.SGDRegressor(n_iter=n_iter, alpha=alpha)
    sgdr.fit(x, y)
    
    return sk.metrics.mean_squared_error(y, sgdr.predict(x))
```

```python
for alpha in [0.1, 1.0, 10.0]:
    plt.title('alpha = {}'.format(alpha))
    plt.plot(range(1, 50), [regression_error(i, alpha) for i in range(1, 50)])
    plt.show()
```
